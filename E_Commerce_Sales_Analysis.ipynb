```python
#for data manipulation
import pandas as pd
#for numerical computation
import numpy as np
#for visulaization
import matplotlib.pyplot as plt
#for advance visulaization
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#read excel file
data=pd.read_excel("Superstore_USA.xlsx")
#5 records print
data.head()
```

```python
#no of rows and columns
data.shape
```

```python
#stats of data
data.describe()
```

```python
#data type
data.info()
```

```python
data.columns
```

```python
#drop unecessry column
data.drop(['Row ID','Discount','Unit Price','Customer ID','Customer Name','Postal Code','Quantity ordered new','Order ID'],axis=1,inplace=True)
```

```python
#check null values
data.isnull().sum()
```

```python
#fill missing values
#for numercial-mean
#for categorical -median
data['Product Base Margin']=data['Product Base Margin'].fillna(data['Product Base Margin'].mean())
```

```python
#check data having null value or not
data.isnull().sum()
```

```python
#check value of order priority
data['Order Priority'].value_counts()
```

```python
#check unique values in order priority
data['Order Priority'].unique()
```

```python
#replace critical with critical
data['Order Priority']=data['Order Priority']=data['Order Priority'].replace('Critical ','Critical')
```

```python
#check unique values in order priority
data['Order Priority'].unique()
```
